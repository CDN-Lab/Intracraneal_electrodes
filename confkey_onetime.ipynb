### One time fix 

We are missing a confidecne column for one participant, we will make this one time fix

```python
import os
import pandas as pd
pd.options.display.max_rows = 999
session_dir = '/Volumes/UCDN/datasets/ICR/raw_data/NIH117/session2'
subj_fn = os.path.join(session_dir,'NIH117_S2_2023-09-05_15h25.39.769.csv')

df = pd.read_csv(subj_fn)

columns = list(df)

conf_source = 'crdm_conf_resp.keys'
conf_destination = 'crdm_confkey'

conf_idx = columns.index(conf_source)

mapped_dict = {'k':'4', 'j':'3', '3':'2', '2':'1', None:None}


df[conf_destination] = df[conf_source].map(mapped_dict)
# [mapped_dict[k] for k in df['crdm_conf_resp.keys']]
# df[['crdm_confkey','crdm_conf_resp.keys']]

print(columns)

columns = columns[:conf_idx] + [conf_destination] + columns[conf_idx:]

print(columns)

df = df.reindex(columns, axis=1)
subj_fn = subj_fn.replace('.csv','_confkey.csv')
print('Saving to : {}'.format(subj_fn))
df.to_csv(subj_fn, index=False)
df
```
